```python
import numpy as np
print("                                           ")
print("               CALCULATOR                  ")
print("                                          ")


def ArrayOperations(n):
    n1=input("enter the arr1 ele with space")
    arr1=np.array([int(i) for i in n1.split()])
    n2=input("enter the arr2 ele with space")
    arr2=np.array([int(i) for i in n2.split()])
    operator=input("""Enter 'all' if u need all the operations
    Enter '+' if u need addition
    Enter '-' if u need subraction
    Enter '*' if u need multiplication
    Enter '/' if u need division
    Enter '//' if u need floor division
    Enter '%' if u need moduleDivision
    Eneter '**' if u need exponensial """)
    print(arr1)
    print(arr2)
    print("What could you like to perform")
    if operator == "all":
      print(f"addition         : {arr1+arr2}")
      print(f"subraction       : {arr1-arr2}")
      print(f"multiplication   : {arr1*arr2}")
      print(f"division         : {arr1/arr2}")
      print(f"floorDivision    : {arr1//arr2}")
      print(f"modulusDivision  : {arr1%arr2}")
      print(f"exponensial      : {arr1**arr2}")
      print("************** END ****************")
    elif operator == '+':
      print(f"addition         : {arr1+arr2}")
    elif operator == '-':
      print(f"subraction       : {arr1-arr2}")
    elif operator == '*':
      print(f"multiplication   : {arr1*arr2}")
    elif operator == '/':
      print(f"division         : {arr1/arr2}")
    elif operator =='//':
      print(f"floorDivision    : {arr1//arr2}")
    elif operator == '%':
      print(f"exponensial      : {arr1**arr2}")

def StatisticalOperation(n):
    n1=input("enter the arr ele with space")
    arr=np.array([int(i) for i in n1.split()])
    operation=input(""" Enter 'all' if u need all the Statistical operation
    Enter 'men' if u need Mean
    Enter 'med' if u need Median
    Enter 'sd' if u need Standard Diviation
    Enter 'var' if u need variance
    Enter 'min' if u need minimum value
    Enter 'max' if u need maximum
    Enter 'sum' if u need sum of the number in array
    Enter 'prd' if u need product of the number in array
    Enter 'sp' if u need shape of  the array chaged
    Enter 'sz' if u need size of the array changed
    Enter 'dt' if u what to know the datatype""")

    if operation == "all":
      print(f"Mean : {(arr).mean()}")
      print(f"Median : {np.median(arr)}")
      print(f"standard diviation : {np.std(arr)}")
      print(f"variance : {np.var(arr)}")
      print(f"minimum value : {min(arr)}")
      print(f"Maximum value : {max(arr)}")
      print(f"sum : {sum(arr)}")
      print(f"product : {np.prod(arr)}")
      print(f"Shape : {np.shape(arr)}")
      print(f"size : {np.size(arr)}")
      print(f"datatype : {arr.dtype}")
    elif operation == "men":
      print(f"Mean : {(arr).mean()}")
    elif operation == 'med':
      print(f"Median : {np.median(arr)}")
    elif operation == 'sd':
      print(f"standard diviation : {np.std(arr)}")
    elif operation == 'var':
      print(f"variance : {np.var(arr)}")
    elif operation == 'min':
      print(f"minimum value : {min(arr)}")
    elif operation == 'max':
      print(f"Maximum value : {max(arr)}")
    elif operation == 'sum':
      print(f"sum : {sum(arr)}")
    elif operation == 'prd':
      print(f"product : {np.prod(arr)}")
    elif operation == 'sp':
      print(f"Shape : {np.shape(arr)}")
    elif operation == 'sz':
      print(f"size : {np.size(arr)}")
    elif operation == 'dt':
      print(f"datatype : {arr.dtype}")
    print("************** END ****************")


def ArrayManu(n):
    n1=input("enter the arr element with space")
    arr=np.array([int(i) for i in n1.split()])
    operator=input("""enter  'res' to reshape the array
    enter 'sor' to sort the array
    enter 'uni' to unique the array
    enter 'finx to get first array
    enter 'linx' to get last array """)
    if operator == 'res':
        ress = int(input("enter the reshape dimension 2 or 3"))
        if ress==2 or 1:
          row=int(input("enter the rowsize"))
          col=int(input("enter the columnsize"))
          try:
            if row*col == arr:
              print(f"reshaping:{arr.reshape(row,col)}")
          except:
            print("Can not reshape")
        else:
          row=int(input("enter the row size"))
          col=int(input("enter the column size"))
          ele=int(input("enter the element size"))
          try:
            if row*col*ele == arr:
              print(f"reshaping:{arr.reshape(row,col,ele)}")
          except:
            print("can not reshape ")
    elif operator =='sor':
      print(f"sort {np.sort(arr)}")
    elif operator == 'uni':
      print(f"unique {np.unique(arr)}")
    elif operator == 'finx':
      print(f"array 1st ele {arr[0]}")
    elif operator == 'linx':
      print(f"array last ele {arr[len(arr)-1]}")

    print("************** END ****************")

def Choose(b):
    n=int(b)
    if n==1:
        ArrayOperations(n)
    elif n==2:
        StatisticalOperation(n)
    elif n==3:
        ArrayManu(n)

while True:
  print("********** Choose anyone to perfome calculation **********")
  print("1.ARITHMETIC OPERATIONS")
  print("2.STATISTICAL OPERATIONS")
  print("3.ARRAY MANUPULATION")
  print("4.EXIT")
  n=int(input())
  if n<4:
      Choose(n)
  else:
      print("********** Exit Completed *********")
      break
```
